```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# reading data from yahoo
from pandas_datareader.data import DataReader
import yfinance as yf
from pandas_datareader import data as pdr

# yf.pdr_override()

# for time stamp
from datetime import datetime, timedelta


# selected securities
tech_list = ['AAPL', 'MSFT', 'AMZN', 'TSLA']


# start and end time of the stock
end = datetime.now()
start = datetime(end.year-1, end.month, end.day)

for stock in tech_list:
    globals()[stock] = yf.download(stock, start, end)
    
company_list = [AAPL, MSFT, AMZN, TSLA]
company_name = ["APPLE", "MICRODOFT", "AMAZON", "TESLA"]


for company, com_name in zip(company_list, company_name):
    company["company_name"] = com_name
    
df = pd.concat(company_list, axis=0)
# df = df.reset_index() # to add date as a column
df.head(10)
```

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Adj Close'].plot()
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {tech_list[i - 1]}")
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    company['Volume'].plot()
    plt.ylabel('Volume')
    plt.xlabel(None)
    plt.title(f"Volume of {tech_list[i - 1]}")
    
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for i, company in enumerate(company_list, 1):
    plt.subplot(2, 2, i)
    
    # Plot the closing price
    ax1 = plt.gca()
    company['Adj Close'].plot(ax=ax1, color='b')
    ax1.set_ylabel('Adj Close', color='b')
    ax1.set_xlabel(None)
    ax1.set_title(f"Closing Price and Volume of {tech_list[i - 1]}")
    
    # Plot the volume on the same subplot with a different y-axis
    ax2 = ax1.twinx()
    company['Volume'].plot(ax=ax2, color='r', alpha=0.6)
    ax2.set_ylabel('Volume', color='r')
    
plt.tight_layout()
plt.show()
```

```python
ma_day = [10, 20, 50]

for ma in ma_day:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Adj Close'].rolling(ma).mean()
        

fig, axes = plt.subplots(nrows=2, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

AAPL[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,0])
axes[0,0].set_title("APPLE")

MSFT[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,1])
axes[0,1].set_title("MICROSOFT")

AMZN[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,0])
axes[1,0].set_title("AMAZON")

TSLA[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,1])
axes[1,1].set_title("TESLA")

fig.tight_layout()
```
