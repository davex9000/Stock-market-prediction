```python
# 1. Getting the earnings (daily)

# things to be corrected:
# 1. scrap data from NASDAQ, yfinance is not accurate


import requests
from bs4 import BeautifulSoup
import datetime

# Today's date
today_date = datetime.date.today().strftime('%Y-%m-%d')

# URL for Yahoo Finance earnings calendar
url = f'https://finance.yahoo.com/calendar/earnings?day={today_date}'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the earnings table (the actual structure may vary)
earnings_table = soup.find('table', {'class': 'W(100%)'})

ticker_today = []
company_name_today = []

# Extract and print company names and tickers
if earnings_table:
    rows = earnings_table.find_all('tr')
    print("Companies with earnings releases today:", today_date)
    for row in rows[1:]:  # Skip the header row
        cols = row.find_all('td')
        ticker = cols[0].text.strip()
        ticker_today.append(ticker)
        company_name = cols[1].text.strip()
        company_name_today.append(company_name)
        print(ticker, company_name)
else:
    print("No earnings releases found for today.")
```

```python
print(ticker_today)
```

```python
# 2. seperating bio-tech and pharma industries
healthcare_companies = []

for i in range(len(ticker_today)):
    ticker_type = yf.Ticker(ticker_today[i])
    try:
        print(ticker_today[i], ":", company_name_today[i], ':', ticker_type.info["sectorDisp"])
        if ticker_type.info["sectorDisp"] == "Healthcare":
            healthcare_companies.append(ticker_today[i])
    except:
      print(ticker_today[i], ":", company_name_today[i], ':', "Not available")
```
